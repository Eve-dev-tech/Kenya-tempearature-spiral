# Generating Kenya Tempatrature Sprial 


 The goal of this project is to create an animated spiral of Kenya's variation in temperature from 1991 to 2016.

Background: Ed Hawkins, a climate scientist, unveiled an animated visualization in 2017 which captivated the world. This visualization showed the deviations of the global average temperature from 1850 to 2017. It was reshared millions of times over Twitter and Facebook and a version of it was even shown at the opening ceremony for the Rio Olympics.

This animation is created with the help of https://www.dataquest.io/blog/climate-temperature-spirals-python/ written by Srini Kadamati.

Historical weather data was retrieved from africa open data https://africaopendata.org/dataset/kenya-climate-data-1991-2016 

The data was collected by the climate knowledge portal by the World Bank.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

months=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(111, projection='3d')

# Create spiral data
theta = np.linspace(0, 2*np.pi, 11)
#r = np.arange(0, 2*np.pi, 2*np.pi/12)
r = np.arange(0, 2*np.pi, 2*np.pi/len(month))
months_angles= np.linspace((np.pi/2)+(2*np.pi),np.pi/2,13)
X = np.outer(r, np.cos(theta))
Y = np.outer(r, np.sin(theta))
Z = np.outer(np.ones(len(r)), np.arange(-2, 3.5, 0.5))

# Plot the spiral
ax1.plot_surface(X, Y, Z, cmap= cm.coolwarm, alpha=0.8)

# Set axis labels and limits
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Temperature')
ax1.set_xlim(-3, 3)
ax1.set_ylim(-3, 3)
ax1.set_zlim(-2, 3.5)

# Set background and foreground colors
fig.set_facecolor("#323331")
ax1.set_facecolor('#000100')

# Add month labels
for i, month in enumerate(months):
    angle = months_angles[i]
    x = (2 + 0.5) * np.cos(angle)
    y = (2 + 0.5) * np.sin(angle)
    z = -2
    ax1.text(x, y, z, month, color="white", fontsize=12, ha="center")

# Add year labels and update spiral in animation
years = temp_data_2['Year'].unique()
lines = []

def update(i):
    # Remove previous year's line
    if lines:
        for line in lines:
            line.remove()
    # Add new year's line and label
    year = years[i]
    #color = plt.cm.viridis(i * 30)
    color = plt.cm.viridis(i / len(years))
    r = temp_data_2[temp_data_2['Year'] == year]['Temparature'] + 2
    X_line = r * np.cos(theta)
    Y_line = r * np.sin(theta)
    Z_line = np.full(len(r), i * 0.5)
    line = ax1.plot(X_line, Y_line, Z_line, c=color)
    label = ax1.text(0, 0, i * 0.5, year, fontsize=20, color="white", ha="center")
    lines.append(line[0])
    return lines + [label]

anim = animation.FuncAnimation(fig, update, frames=len(years), interval=10)

ffmpeg_writer = animation.FFMpegWriter()
anim.save("Spiral.gif", writer='pillow', fps=4, dpi=100)
#anim.save("Spiral.gif", writer=ffmpeg_writer, fps=4, dpi=100)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.animation import FuncAnimation, FFMpegWriter

months=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(111, projection='3d')

# Create spiral data
theta = np.linspace(0, 2*np.pi, 11)
#r = np.arange(0, 2*np.pi, 2*np.pi/len(months))
r = np.arange(0, 2*np.pi, 2*np.pi/(len(months)-1))
months_angles = np.linspace((np.pi/2)+(2*np.pi),np.pi/2,len(months)+1)
X = np.outer(r, np.cos(theta))
Y = np.outer(r, np.sin(theta))
Z = np.outer(np.ones(len(r)), np.arange(-2, 3.5, 0.5))

# Plot the spiral
ax1.plot_surface(X, Y, Z, cmap=cm.coolwarm, alpha=0.8)

# Set axis labels and limits
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Temparature')
ax1.set_xlim(-3, 3)
ax1.set_ylim(-3, 3)
ax1.set_zlim(-2, 3.5)

# Set background and foreground colors
fig.set_facecolor("#323331")
ax1.set_facecolor('#000100')

# Add month labels
for i, month in enumerate(months):
    angle = months_angles[i]
    x = (2 + 0.5) * np.cos(angle)
    y = (2 + 0.5) * np.sin(angle)
    z = -2
    ax1.text(x, y, z, month, color="white", fontsize=12, ha="center")

# Add year labels and update spiral in animation
years = temp_data_2['Year'].unique()
lines = []

def update(i):
    # Remove previous year's line
    if lines:
        for line in lines:
            line.remove()
    # Add new year's line and label
    year = years[i]
    color = plt.cm.viridis(i / len(years))
    r = temp_data_2[temp_data_2['Year'] == year]['Temparature'] + 2
    X_line = r * np.cos(theta)
    Y_line = r * np.sin(theta)
    #Z_line = np.full(len(r), i * 0.5)
    Z_line = np.full(len(r), float(i) * 0.5)

    line = ax1.plot(X_line, Y_line, Z_line, c=color)
    label = ax1.text(0, 0, i * 0.5, year, fontsize=20, color="white", ha="center")
    lines.append(line[0])
    return lines + [label]

anim = FuncAnimation(fig, update, frames=len(years), interval=10)

ffmpeg_writer = FFMpegWriter()
anim.save("Spiral.gif", writer='pillow', fps=4, dpi=100)
```
